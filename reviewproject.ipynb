## Customer Feedback Data Analysis
This project analyzes customer feedback based on product purchases, demographics, and ratings. 
The goal is to uncover insights that can help businesses improve their online shopping experience.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel("Updated_Customer_Feedback_Dataset.xlsx")

# Display first few rows
df.head()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))
```

## Most Purchased Products
This bar chart shows the most frequently purchased products. Understanding which products are most popular can help businesses focus on top-selling items.

```python
plt.figure(figsize=(10, 5))
ax = sns.countplot(data=df, y='Product Purchased', order=df['Product Purchased'].value_counts().index, palette='magma')
plt.xlabel("Number of Purchases")
plt.ylabel("Product")
plt.title("Most Purchased Products")
for container in ax.containers:
    ax.bar_label(container, fmt='%d')
plt.show()
```

## Number of Purchases per Platform
This bar chart displays the total purchases per platform, helping businesses assess which platform is the most popular among customers.

```python
plt.figure(figsize=(10, 5))
ax = sns.countplot(data=df, y='Purchase Platform', order=df['Purchase Platform'].value_counts().index, palette='viridis')
plt.xlabel("Count")
plt.ylabel("Purchase Platform")
plt.title("Purchase Platform Usage Distribution")
for container in ax.containers:
    ax.bar_label(container, fmt='%d')
plt.show()
```

## Gender Distribution
This pie chart represents the percentage of male, female, and other gender customers. This helps businesses understand their audience better.

```python
plt.figure(figsize=(6, 6))
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'pink', 'gray'], startangle=90, explode=[0.05, 0.05, 0.05])
plt.title("Gender Distribution")
plt.ylabel('')
plt.show()
```

## Customer Age Distribution
This line chart shows how customers are distributed across different age groups. It helps businesses understand which age group shops the most.

```python
plt.figure(figsize=(8, 5))
age_counts = df["Age"].value_counts().sort_index()
plt.plot(age_counts.index, age_counts.values, marker='o', linestyle='-', color='b')
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.title("Customer Distribution by Age")
plt.show()
```

## Feature Correlation Heatmap
This heatmap displays the correlation between numerical features like Age and Rating. Darker colors indicate stronger correlations.
It helps identify relationships between different variables in customer feedback.

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

## Customer Distribution by Location
This pie chart shows how customers are distributed by location, helping businesses understand regional purchasing trends.

```python
plt.figure(figsize=(6, 6))
df["Location"].value_counts().plot.pie(autopct="%1.1f%%", cmap="coolwarm", startangle=90)
plt.title("Customer Distribution by Location")
plt.ylabel("")
plt.show()
```

## Customer Type Distribution
This donut chart represents the percentage of different customer types, helping businesses target their ideal audience effectively.

```python
plt.figure(figsize=(6, 6))
customer_type_counts = df["Customer Type"].value_counts()
plt.pie(customer_type_counts, labels=customer_type_counts.index, autopct='%1.1f%%', colors=sns.color_palette("coolwarm"), wedgeprops={'edgecolor': 'white'}, startangle=90, pctdistance=0.85)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title("Customer Type Distribution")
plt.show()
```

## Return Status Distribution
This bar chart shows the distribution of returned products. It helps businesses identify the return trends and improve product quality

```python
plt.figure(figsize=(8, 5))
return_status_counts = df["Return Status"].value_counts()
ax = return_status_counts.plot(kind="bar", color=['#FF9999', '#66B3FF', '#99FF99', '#FFCC99'])
plt.title("Return Status Distribution")
plt.xlabel("Return Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
for container in ax.containers:
    ax.bar_label(container, fmt='%d')
plt.show()
```
